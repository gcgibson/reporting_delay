```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

D= 10
ps = []
cov_1 = 1*np.eye(D)

cov_2 = 1*np.eye(D)


catch_up = False
mean_vec_1 = np.ones(D)
mean_vec_1[5:D] = 0.0001
mean_vec_1[0]= 0
mean_vec_1 = 1*mean_vec_1
mean_vec_1[-1] = 0

mean_vec_2 = np.ones(D)
mean_vec_2[5:D] = 0.0001
mean_vec_2[0]= 0
mean_vec_2 = .1*mean_vec_2
mean_vec_2[-1] = 0

for i in range(100):
            
    
  
    
    alphas_1 = np.random.multivariate_normal(mean_vec_1,cov_1)
    alphas_2 = np.random.multivariate_normal(mean_vec_2,cov_2)

    A = np.eye(len(alphas_1))
    A[1,3] = 1
    A[2,1] = 1
    A[3,2] = 1
    A[1,1] = 0
    A[2,2] = 0
    A[3,3] = 0
    A[4,4] = 1
    A = np.transpose(A)
    
    B = np.eye(D)
    mean_vec_1 = .99999*np.dot(A,alphas_1)
    
    mean_vec_2 = .99999*np.dot(B,alphas_2)

    
    
    if np.random.binomial(1,.65) == 1:
        #print (np.exp(alphas_1))
        ps.append(np.random.dirichlet(np.exp(alphas_1)))
        
    else:
        #print (np.exp(alphas_2))
        ps.append(np.random.dirichlet(np.exp(alphas_2)))
        
    

plt.figure(figsize=(8,8))
plt.ylabel("Bi-week")
plt.xlabel("d")
#plt.xticks(0,199,1)
plt.imshow(ps,aspect='auto',cmap=cm.Blues)
plt.show()
```
